```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("M16-Amazon-Challenge").config("spark.driver.extraClassPath","/content/postgresql-42.2.16.jar").getOrCreate()
```

```python
from pyspark import SparkFiles
url = "https://whaidaryar-bucket.s3.amazonaws.com/amazon_reviews_us_Home_Entertainment_v1_00.tsv"
spark.sparkContext.addFile(url)
df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get(""), sep="\t", header=True, inferSchema=True)
df.show()
```

```python
vine_df = df.select(["review_id", "star_rating", "helpful_votes", "total_votes", "vine", "verified_purchase"])
vine_df.show()
```

```python
total_vote_count = vine_df.filter("total_votes >= 20")

total_vote_count.show()
```

```python
most_helpful_votes = total_vote_count.filter(total_vote_count["helpful_votes"]/total_vote_count["total_votes"]>=0.5)
most_helpful_votes.show()
```

```python
vine_paid_df = most_helpful_votes.filter("vine == 'Y'")
vine_paid_df.show()
```

```python
vine_free_df = most_helpful_votes.filter("vine == 'N'")
vine_free_df.show()
```

```python
# Total Number of reviews for
most_helpful_votes.count()
```

```python
Five_Star_Rating_df = most_helpful_votes.filter(most_helpful_votes["star_rating"]== 5)
Five_Star_Rating_df.show()
```

### **Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).**

```python
#Total Paid Reviews
total_paid_reviews = (vine_paid_df).count()
total_paid_reviews
```

```python
# Total unpaid reviews
total_free_reviews = (vine_free_df).count()
total_free_reviews
```

```python
# All 5-Star Reviews
FiveStar_Reviews = (Five_Star_Rating_df).count()
FiveStar_Reviews
```

```python
# Paid 5-Star Reviews
paid_fivestar_reviews = vine_paid_df.filter(vine_paid_df.star_rating == 5).count()
paid_fivestar_reviews
```

```python
# un-paid 5-Star Reviews

unpaid_fivestar_reviews = vine_free_df.filter(vine_free_df.star_rating == 5).count()
unpaid_fivestar_reviews
```

```python
# Percentage for paid 5-Star Reviews
percentage_paid_fivestar = (paid_fivestar_reviews / total_paid_reviews) * 100 
percentage_paid_fivestar
```

```python
# Percentage for un-paid 5-Star Reviews

percentage_unpaid_fivestar = (unpaid_fivestar_reviews / total_free_reviews) * 100
percentage_unpaid_fivestar
```
